```python
class Employee:
    def __init__(self, name, salary, department):
        self.name = name
        self.salary = salary
        self.department = department
        self.leave_days = 0
        self.projects = []

    # 1ï¸âƒ£ Show employee details
    def show_info(self):
        print(f"Name: {self.name}, Department: {self.department}, Salary: Rs.{self.salary}")

    # 2ï¸âƒ£ Give a raise
    def give_raise(self, amount):
        self.salary += amount
        print(f"{self.name}'s salary increased by Rs.{amount}. New salary: Rs.{self.salary}")

    # 3ï¸âƒ£ Deduct tax
    def deduct_tax(self, percent):
        tax = self.salary * (percent / 100)
        self.salary -= tax
        print(f"Rs.{tax} tax deducted. Remaining salary: Rs.{self.salary}")

    # 4ï¸âƒ£ Add overtime pay
    def add_overtime(self, hours):
        rate = 500
        self.salary += hours * rate
        print(f"{self.name} worked {hours} extra hours. Salary is now Rs.{self.salary}")

    # 5ï¸âƒ£ Apply for leave
    def apply_leave(self, days):
        self.leave_days += days
        print(f"{self.name} applied for {days} days leave. Total leave taken: {self.leave_days}")

    # 6ï¸âƒ£ Assign department
    def change_department(self, new_dept):
        old = self.department
        self.department = new_dept
        print(f"{self.name} moved from {old} to {new_dept} department.")

    # 7ï¸âƒ£ Assign a project
    def add_project(self, project_name):
        self.projects.append(project_name)
        print(f"{self.name} is assigned to project: {project_name}")

    # 8ï¸âƒ£ Show all assigned projects
    def show_projects(self):
        if self.projects:
            print(f"{self.name}'s Projects: {', '.join(self.projects)}")
        else:
            print(f"{self.name} has no projects assigned yet.")

    # 9ï¸âƒ£ Annual salary calculation
    def annual_salary(self):
        yearly = self.salary * 12
        print(f"{self.name}'s annual salary is Rs.{yearly}")
        return yearly

    # ðŸ”Ÿ Appreciation message
    def appreciate(self):
        print(f"ðŸŒŸ Excellent work, {self.name}! Keep shining in the {self.department} department!")

# ------------------------------------
# Example Usage
emp = Employee("Ali", 50000, "IT")

emp.show_info()
emp.give_raise(5000)
emp.deduct_tax(5)
emp.add_overtime(3)
emp.apply_leave(2)
emp.change_department("Finance")
emp.add_project("Budget Automation")
emp.add_project("Annual Report System")
emp.show_projects()
emp.annual_salary()
emp.appreciate()
```

## PROTECTED

```python
class Person:
    def __init__(self,name,age):
        self._age = age
        self.name = name
class Student(Person):
    def display_age(self):
        print("Age",self._age)

obj = Student("Memoona",22)
obj.display_age()
print(obj._age)
```
