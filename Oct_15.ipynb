## Stacking and Splitting Array in Numpy

```python

import random

actions = ["greet", "jump", "sleep", "dance", "sing", "run"]

def ai_action():
    action = random.choice(actions)
    return action

print("\nRandom AI Agent Behavior:")
for step in range(10):
    print(f"Step {step+1}: AI decides to {ai_action()}")
```
