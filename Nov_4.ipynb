## using code make folder in a folder then sone file

```python
import os

# Define the folder structure
structure = {
    "B": {
        "C": {
            "E": ["h1.txt"],
            "F": ["h2.bib"]
        },
        "D": ["hello.py", "hi.ipynb"]
    }
}

# Function to create folders and files recursively
def create_structure(base_path, structure):
    for key, value in structure.items():
        folder_path = os.path.join(base_path, key)
        os.makedirs(folder_path, exist_ok=True)
        if isinstance(value, dict):
            create_structure(folder_path, value)
        elif isinstance(value, list):
            for file_name in value:
                open(os.path.join(folder_path, file_name), 'w').close()

# Run the function from the current directory
create_structure(".", structure)

print("✅ Folder and file structure created successfully!")
```
