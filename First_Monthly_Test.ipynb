## 1️⃣ Data Types and Conditions

```python
student_name = "Aarav"
marks = [78, 85, 62, 90, 88]
average = sum(marks) / len(marks)

grade = "Pass" if average >= 70 else "Fail"
print(f"Student: {student_name}, Average: {average:.2f}, Result: {grade}")
```

## 2️⃣ Loops and Functions

```python
def calculate_percentage(marks_list):
    total = sum(marks_list)
    return [(m / total) * 100 for m in marks_list]

percentages = calculate_percentage(marks)
print("Percentages per subject:", [round(p, 2) for p in percentages])
```

## 3️⃣ NumPy Operations

```python
import numpy as np

np_marks = np.array(marks)
scaled_marks = np_marks * 1.1
print("Scaled Marks (with bonus):", np.round(scaled_marks, 1))

above_85 = np_marks[np_marks > 85]
print("Marks above 85:", above_85)
```

## 4️⃣ Pandas DataFrame

```python
import pandas as pd

data = {
    "Subject": ["Math", "Science", "English", "History", "Art"],
    "Marks": marks,
    "Scaled Marks": scaled_marks
}

df = pd.DataFrame(data)
df["Pass/Fail"] = ["Pass" if m >= 70 else "Fail" for m in df["Marks"]]
print("\nDataFrame:\n", df)

avg_scaled = df["Scaled Marks"].mean()
print(f"\nAverage Scaled Mark: {avg_scaled:.2f}")
```

## 5️⃣ Matplotlib Visualization

```python
import matplotlib.pyplot as plt

plt.bar(df["Subject"], df["Scaled Marks"], color="skyblue", label="Scaled Marks")
plt.axhline(y=avg_scaled, color="red", linestyle="--", label="Average")
plt.title(f"{student_name}'s Performance")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
