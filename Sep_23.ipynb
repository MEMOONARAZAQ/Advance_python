## Positional Argument (*args)

```python
def count_microbes(*args):
    """Takes any number of microbial species and prints a summary."""
    print("Number of microbial species found:", len(args))
    print("Species detected:")
    for species in args:
        print("-", species)

# Example call:
count_microbes("Lactobacillus", "Geotrichum candidum", "Bifidobacterium", "Saccharomyces")
```

```python
def colony_count(*args):
    """Takes CFU/ml counts of different samples and gives total + average."""
    total = sum(args)
    avg = total / len(args)
    print("Total colonies counted:", total)
    print("Average colonies per sample:", avg)

# Example:
colony_count(120, 250, 300, 450)
```

```python
def antibiotic_sensitivity(*args):
    """List all antibiotics a strain is sensitive to."""
    print("Sensitive to the following antibiotics:")
    for drug in args:
        print("✔", drug)

# Example:
antibiotic_sensitivity("Amoxicillin", "Ciprofloxacin", "Erythromycin")
```

```python
def species_ph(*args):
    """Takes species name and its optimum pH in pairs."""
    print("Species and their optimum pH:")
    for i in range(0, len(args), 2):
        print(f"{args[i]}: optimum pH = {args[i+1]}")

# Example:
species_ph("Lactobacillus", 5.5, "Geotrichum candidum", 6.5, "Saccharomyces", 4.5)
```

```python
def sample_batch(*args):
    """Show the samples combined into one batch test."""
    print("Samples included in this batch test:")
    for sdfgh in args:
        print(sdfgh)

# Example call
sample_batch("Milk sample 1", "Yogurt sample 2", "Cheese sample 3")
```

```python
def growth_temps(*args):
    """Takes species and their temperatures in pairs and prints them."""
    print("Species and their optimum temperatures:")
    for i in range(0, len(args), 2):
        print(f"{args[i]} grows best at {args[i+1]}°C")

# Example call
growth_temps("Lactobacillus", 37, "Geotrichum candidum", 28, "Saccharomyces", 30)
```
