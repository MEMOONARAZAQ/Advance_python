## Creating Sub Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.subplot(3, 4, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Subplot 1")
plt.subplot(3, 4, 3)
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Subplot 2")
plt.subplot(3, 4, 6)
plt.bar([1, 2, 3], [7, 8, 9])
plt.title("Subplot 3")
plt.subplot(3, 4, 8)
plt.plot([1, 2, 3], [9, 8, 7])
plt.title("Subplot 4")
plt.subplot(3, 4, 11)
plt.bar([1, 2, 3], [5, 5, 5])
plt.title("Subplot 5")
plt.tight_layout()
plt.show()
```

## Histogram of Age

```python
df = pd.read_excel("data.xlsx")
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Age [years]"], bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.7)
plt.show()
```

## Pie charts for each disease column

```python
disease_columns = [col for col in df.columns if "disease" in col.lower()]

for col in disease_columns:
    counts = df[col].value_counts(normalize=True) * 100
    plt.figure(figsize=(4, 4))
    plt.pie(
        counts,
        labels=[f"{i}" for i in counts.index],
        autopct="%1.1f%%",
        colors=['yellow', 'lightcoral']
    )
    plt.title(f"{col} (%)")
    plt.show()
```

## Pie chart for Socioeconomic status

```python
plt.figure(figsize=(5, 5))
df["Socioeconomic status"].value_counts(normalize=True).plot.pie(
    autopct='%1.1f%%',
    colors=['gold', 'skyblue', 'lightcoral'],
    ylabel=''
)
plt.title("Socioeconomic Status (%)")
plt.show()
```

## Pie chart for Sex

```python
smoking_cols = [col for col in df.columns if "smok" in col.lower()]
if smoking_cols:
    smoking_col = smoking_cols[0]
    plt.figure(figsize=(5, 5))
    df[smoking_col].value_counts(normalize=True).plot.pie(
        autopct='%1.1f%%',
        colors=['green', 'red', 'lightgray'],
        ylabel=''
    )
    plt.title("Smoking Status (%)")
    plt.show()
else:
    print("No smoking-related column found in the dataset.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Select all columns that contain the word 'disease'
disease_columns = [col for col in df.columns if "disease" in col.lower()]

# Calculate how many people have each disease (sum of 1's in each column)
disease_totals = df[disease_columns].sum()

# Convert to percentage of total participants
disease_percentages = (disease_totals / len(df)) * 100

# Make one pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    disease_percentages,
    labels=disease_percentages.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Pastel1.colors
)
plt.title("Overall Disease Distribution (%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ensure df is loaded and normalized
# df = pd.read_excel("data.xlsx")   # uncomment if necessary
print("Columns available:", list(df.columns))

# Identify disease columns and extra categorical columns
disease_columns = [col for col in df.columns if "disease" in col.lower()]
extra_charts = ['Faculty', 'Socioeconomic_Status', 'Sex', 'Smoking']

# compute grid size
total_plots = 1 + len(disease_columns) + len(extra_charts)
cols = 3
rows = (total_plots // cols) + (total_plots % cols > 0)

fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 5 * rows))
axes = axes.flatten()

colormaps = ['Set3', 'Pastel1', 'Accent', 'Dark2', 'tab10', 'Paired', 'Set2', 'tab20']

# 1) Histogram for age  try common column names safely
age_col = None
for candidate in ['Age', 'Age [years]', 'Age [yrs]']:
    if candidate in df.columns:
        age_col = candidate
        break
if age_col is None:
    raise KeyError("No Age column found. Available columns: " + ", ".join(df.columns))
axes[0].hist(pd.to_numeric(df[age_col], errors='coerce').dropna(), bins=10,
             color='skyblue', edgecolor='black')
axes[0].set_title("Age Distribution", fontsize=12, fontweight='bold')
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")
axes[0].grid(axis='y', alpha=0.7)

# 2) Disease pie charts (start at index 1)
idx = 1
for i, col in enumerate(disease_columns):
    counts = df[col].value_counts(normalize=True) * 100
    cmap = plt.cm.get_cmap(colormaps[i % len(colormaps)], max(3, len(counts)))
    colors = [cmap(j) for j in range(len(counts))]
    axes[idx].pie(counts, labels=[str(label) for label in counts.index],
                  autopct="%1.1f%%", colors=colors, textprops={'fontsize': 9})
    axes[idx].set_title(f"{col} (%)", fontsize=12, fontweight='bold')
    idx += 1

# 3) Extra categorical columns after disease charts
start_index = idx
for j, col in enumerate(extra_charts, start=start_index):
    if col in df.columns:
        counts = df[col].value_counts(normalize=True) * 100
        cmap = plt.cm.get_cmap(colormaps[(j - start_index) % len(colormaps)], max(3, len(counts)))
        colors = [cmap(k) for k in range(len(counts))]
        axes[j].pie(counts, labels=[str(label) for label in counts.index],
                    autopct="%1.1f%%", colors=colors, textprops={'fontsize': 9})
        axes[j].set_title(f"{col} (%)", fontsize=12, fontweight='bold')
    else:
        # If column doesn't exist, remove that axis to keep layout clean
        fig.delaxes(axes[j])

# remove any remaining empty axes (in case total_plots < rows*cols)
for rem in range(idx + len(extra_charts), len(axes)):
    try:
        fig.delaxes(axes[rem])
    except Exception:
        pass

plt.tight_layout()
plt.show()
```
