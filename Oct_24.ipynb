## DNA Sequence Analyzer Project — Description

Objective:
The project analyzes DNA sequences to calculate their GC content (the percentage of Guanine and Cytosine bases). This helps in understanding the stability and composition of the DNA.

Data Loading (FASTA files):
DNA sequences are stored in FASTA format. The program reads these files and extracts sequence IDs and their nucleotide sequences using Python.

Data Cleaning:
Sequences are cleaned by removing unwanted characters and converting all bases to uppercase (A, T, G, C only).

GC Content Calculation:
Using loops and conditions, the program counts how many G and C bases are present and calculates the GC percentage for each sequence.

Categorization (Ternary Operator):
Each sequence is labeled as “High GC” or “Low GC” depending on whether its GC percentage is above or below 50%.

NumPy Computation:
NumPy arrays are used to calculate statistical summaries like the mean, maximum, and minimum GC content.

List Comprehension:
Summaries of each sequence are created using list comprehensions for faster and cleaner results.

Functions and Modularization:
All steps are written inside Python functions and saved in a module (dna_analyzer.py) so the code is reusable and well-organized.

Output:
The final program displays each sequence with its GC content, category (High/Low), and an overall summary of GC statistics.
