## Examples

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound
    
    def make_sound(self):
        print(f"{self.name} says {self.sound}")

cat = Animal("Cat", "Meow")
dog = Animal("Dog", "Woof")
cat.make_sound()
dog.make_sound()
```

```python
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def display_info(self):
        print(f"Name: {self.name}, Roll No: {self.roll_no}")

student1 = Student("Memoona", 101)
student1.display_info()
```

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def description(self):
        print(f"'{self.title}' by {self.author} has {self.pages} pages.")

book1 = Book("Pride and Prejudice", "Jane Austen", 432)
book1.description()
```

```python
class Laptop:
    def __init__(self, brand, ram, price):
        self.brand = brand
        self.ram = ram
        self.price = price

    def details(self):
        print(f"Laptop Brand: {self.brand}, RAM: {self.ram}GB, Price: ${self.price}")

laptop1 = Laptop("HP", 16, 850)
laptop1.details()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} vehicle started!")

class Car(Vehicle):
    def __init__(self, brand, color):
        super().__init__(brand)
        self.color = color

    def show(self):
        print(f"{self.color} {self.brand} is ready to go!")

my_car = Car("Honda", "Blue")
my_car.start()
my_car.show()
```
