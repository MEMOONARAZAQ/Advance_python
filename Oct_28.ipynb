## Creating piecharts in Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
```

```python
plt.pie( sizes, labels=labels, autopct='%1.1f%%', startangle=140 )
plt.show()
```

```python
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    explode=(0.1, 0, 0, 0),  # explode 1st slice
shadow=True)
plt.title("Pie Chart Example")
plt.show()
```

```python
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    explode=(0, 0, 0.1, 0),  # explode 1st slice
shadow=True)
plt.title("Pie Chart Example")
plt.show()
```

```python
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    explode=(0, 0, 0, 0.1),  # explode 1st slice
shadow=True)
plt.title("Pie Chart Example")
plt.show()
```

```python
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    explode=(0, 0, 0, 0.5),  # explode 1st slice
shadow=True)
plt.title("Pie Chart Example")
plt.show()
```

## Advance 

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 20, 20]
explode = (0.1, 0.05, 0, 0)  # slightly explode first two slices
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

# Plot pie chart
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    shadow=True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'color': 'black', 'fontsize': 12}
)

# Add center circle for donut look
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Add title
plt.title(" Fruit Popularity (Advanced Pie Chart)", fontsize=16, weight='bold')

# Add legend
plt.legend(wedges, labels, title="Fruits", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Beautify layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Cupcakes ', 'Cookies ', 'Donuts ', 'Brownies ']
sizes = [25, 30, 20, 25]
colors = plt.cm.rainbow(np.linspace(0, 1, len(labels)))  # colorful gradient
explode = (0.1, 0.08, 0.06, 0.04)  # all slices slightly explode

# Create figure
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

# Main pie chart
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct=lambda p: f'{p:.1f}%\n({int(p * sum(sizes) / 100)} pcs)',
    startangle=120,
    colors=colors,
    pctdistance=0.75,
    textprops={'fontsize': 11, 'color': 'white', 'weight': 'bold'},
    wedgeprops={'linewidth': 2, 'edgecolor': 'white', 'alpha': 0.9},
    shadow=True
)

# Add center circle (donut hole)
centre_circle = plt.Circle((0, 0), 0.55, fc='black', alpha=0.15)
fig.gca().add_artist(centre_circle)

# Add fun title
plt.title(" Sweet Treat Popularity Chart ", fontsize=18, fontweight='bold', color='#ff69b4')

# Add cute annotations
for i, w in enumerate(wedges):
    angle = (w.theta2 - w.theta1)/2. + w.theta1
    x = np.cos(np.deg2rad(angle))
    y = np.sin(np.deg2rad(angle))
    ax.text(x*1.3, y*1.3, f"{labels[i].split()[0]} ", ha='center', va='center', fontsize=10, rotation=angle)

# Add stylish legend
plt.legend(wedges, labels, title="Desserts", loc="upper left", bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)  # for repeatable results
x = np.random.randint(1, 11, 100)     # x values between 1 and 10
y = np.random.randint(90, 101, 100)   # y values between 90 and 100

# Create custom colors based on y values (cool gradient look)
colors = plt.cm.cool(np.linspace(0, 1, len(x)))

# Plot
plt.figure(figsize=(9, 6))
plt.scatter(
    x, y,
    c=colors,
    marker='o',
    s=np.random.randint(50, 250, 100),  # random point sizes
    alpha=0.7,
    edgecolors='white',
    linewidths=1.5
)

# Add titles and labels
plt.title(" Fun Scatter Plot: Randomized Data", fontsize=16, weight='bold')
plt.xlabel("X (110 range)", fontsize=12)
plt.ylabel("Y (90100 range)", fontsize=12)

# Add a background grid
plt.grid(True, linestyle='--', alpha=0.5)

# Add a colorbar just for fun
sm = plt.cm.ScalarMappable(cmap='cool')

# Tight layout and show
plt.tight_layout()
plt.show()
```
