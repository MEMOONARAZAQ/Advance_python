```python
def analyze_growth(organism, media_list):
    print(f"Microorganism: {organism}")
    if not media_list:  # check if list is empty
        print("No media provided.")
    else:
        print("Growth media used:")
        for i, medium in enumerate(media_list, start=1):
            print(f"{i}. {medium}")

# Example usage:
analyze_growth("Geotrichum candidum",
               ["MRS broth", "Cheese whey medium", "Yogurt starter medium"])

# If no media:
analyze_growth("Lactobacillus plantarum", [])
```
